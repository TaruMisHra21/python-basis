1. What is the difference between a function and a method in Python?
- Function in Python

A function is a block of reusable code that performs a specific task.

It is defined independently using def keyword.

It can be called directly by its name.

Example:

    def greet(name):

    return f"Hello, {name}!"

    print(greet("Taru"))   # Calling function directly

 Method in Python

A method is also a function, but it is associated with an object (class instance).

It is defined inside a class.

It can only be called using the object (or class).

Example:
    class Person:
    def greet(self, name):
    return f"Hello, {name}!"

    p = Person()
    print(p.greet("Taru"))   # Calling method through object

2. Explain the concept of function arguments and parameters in Python.
- Parameters
Parameters are the variables defined inside the function header.

They act as placeholders for the values a function expects.

Example:

5.  What are iterators in Python and how do they differ from iterables?
1. Iterable

An iterable is any Python object that can return an iterator using the iter() function.

Examples: lists, tuples, strings, sets, dictionaries, range objects.

Iterable objects implement the __iter__() method.

They do not "remember" their state (they just produce a new iterator when asked).

Example:
    my_list = [1, 2, 3]
    print(iter(my_list))   # gives an iterator object

2. Iterator

An iterator is an object that represents a stream of data, returning one element at a time.

It implements two special methods:

__iter__() ‚Üí returns the iterator object itself.

__next__() ‚Üí returns the next element, and raises StopIteration when no more items exist.

Iterators remember their state ‚Üí once exhausted, they cannot be reused unless recreated.

Example:
    my_list = [1, 2, 3]
    
    it = iter(my_list)  # create an iterator

    print(next(it))  # 1

    print(next(it))  # 2
    
    print(next(it))  # 3

# next(it) now will raise StopIteration

6.  Explain the concept of generators in Python and how they are defined.
- What are Generators?

A generator is a special type of iterator in Python that allows you to produce values lazily (one at a time, only when needed) instead of creating and storing the entire sequence in memory.

They are memory-efficient (don‚Äôt store all items at once).

They‚Äôre created either by:

Generator functions (using yield)

Generator expressions (like list comprehensions but with ()

1. Generator Functions

Defined like a normal function, but instead of return, you use yield.

Each time the generator function is called, it resumes from the last yield point.

Example:
    def count_up_to(n):
    i = 1
    while i <= n:
    yield i   # pause here and return value
    i += 1

    gen = count_up_to(3)
    
    print(next(gen))  # 1
    
    print(next(gen))  # 2
    
    print(next(gen))  # 3

# next(gen) ‚Üí raises StopIteration

Here, count_up_to is a generator function that produces numbers on demand.

2. Generator Expressions

They‚Äôre like list comprehensions but with () instead of [].

Example:
    
    squares = (x*x for x in range(5))

    print(next(squares))  # 0

    print(next(squares))  # 1

    print(next(squares))  # 4

This creates a generator object instead of a list.

Why Use Generators?

Memory efficiency: They don‚Äôt store all values in memory (useful for large datasets).

Performance: Generate values on-the-fly.

Readable code: Cleaner than manually writing iterator classes.

7. Advantages of Generators Over Regular Functions
1. Memory Efficiency

A regular function that returns a list builds the entire list in memory before returning it.

A generator produces values one at a time (lazy evaluation), so it doesn‚Äôt need to store everything.

Example:

    def big_list(n):
    
    return [i for i in range(n)]  # uses memory for all n items

    def big_gen(n):
    
    for i in range(n):
    
    yield i  # yields one item at a time

    big_list(1_000_000) ‚Üí consumes a lot of RAM üñ•Ô∏è

    big_gen(1_000_000) ‚Üí consumes almost no extra memory.

2. Lazy Evaluation (On-Demand Computation)

Generators compute the next value only when needed.

Useful when working with infinite sequences or very large datasets.
def infinite_numbers():
    
    i = 1
    
    while True:
    
    yield i
    
    i += 1

    nums = infinite_numbers()

    print(next(nums))  # 1

    print(next(nums))  # 2

# keeps going infinitely...

3. Better Performance for Large Data

Since values are generated on demand, generators can be faster in cases where you don‚Äôt actually need all results at once.

Example: Iterating through a large log file line by line with a generator is faster and lighter than loading the whole file into memory.

4. Cleaner and Simpler Code

Instead of writing an iterator class with __iter__() and __next__(), you just use yield.

Example:
# Without generator (iterator class)

    class Counter:
  
    def __init__(self, n):
  
    self.n = n
  
    self.current = 0
    
    def __iter__(self):
   
    return self
    
    def __next__(self):
    
    if self.current < self.n:
    
    self.current += 1
    
    return self.current
    
    else:
    
    raise StopIteration

    # With generator
    
    def counter(n):
    
    for i in range(1, n+1):
    
    yield i

Generator code is much shorter and easier to understand.

5. Chaining & Composability

Generators can be easily chained together to create pipelines of data processing.

Example:
    
    numbers = (i for i in range(10))     # generator
    
    squares = (x*x for x in numbers)     # another generator
    
    evens = (x for x in squares if x%2==0)

    print(list(evens))  # [0, 4, 16, 36, 64]

8.  Explain the purpose and usage of the `map()` function in Python.
- The map() function is a built-in Python function used to apply a given function to each item of an iterable (like list, tuple, set) and returns a map object (which is an iterator).

 Syntax:
    map(function, iterable, ...)

function ‚Üí The function to apply.

iterable ‚Üí One or more iterable(s) (list, tuple, etc.).

Purpose of map()

To avoid explicit loops when applying a function to all items in an iterable.

Makes code cleaner and more readable.

Works faster in many cases than using manual for loops.

Example 1: Square all numbers
    
    numbers = [1, 2, 3, 4, 5]

    def square(x):
    
    return x*x

    result = map(square, numbers)

    print(list(result))   # [1, 4, 9, 16, 25]
    
    square() is applied to each element in numbers.

Example 2: Using lambda with map

Instead of defining a function, we can use a lambda expression
    
    numbers = [1, 2, 3, 4, 5]
    
    result = map(lambda x: x*x, numbers)

    print(list(result))   # [1, 4, 9, 16, 25]

Example 3: Multiple Iterables

If you pass multiple iterables, map() applies the function to elements from all iterables in parallel.

    a = [1, 2, 3]
    
    b = [4, 5, 6]

    result = map(lambda x, y: x + y, a, b)
    
    print(list(result))   # [5, 7, 9]

Stops when the shortest iterable is exhausted.

Example 4: Convert strings to integers

    strings = ["1", "2", "3", "4"]

    result = map(int, strings)
    
    print(list(result))   # [1, 2, 3, 4]

When to Use map()?

When you want to apply a function to every element of an iterable.

When using lambda functions for quick transformations.

When working with multiple iterables together.

10. . What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
- 1. map()
Purpose: Applies a function to every element of an iterable and returns a new iterator.

Syntax:
    map(function, iterable, ...)

Example:
    
    numbers = [1, 2, 3, 4]

    squared = map(lambda x: x*x, numbers)
    
    print(list(squared))   # [1, 4, 9, 16]

Use Case: Transforming elements (e.g., converting strings to int, squaring numbers).

2. filter()
Purpose: Filters elements of an iterable based on a boolean condition (function returns True or False).

Syntax:
    
    filter(function, iterable)

Example:

    numbers = [1, 2, 3, 4, 5, 6]

    evens = filter(lambda x: x % 2 == 0, numbers)

    print(list(evens))   # [2, 4, 6]

Use Case: Selecting a subset of data (e.g., filter even numbers, remove empty strings).

3. reduce()
Purpose: Repeatedly applies a function to the elements of an iterable, reducing it to a single cumulative value.

Unlike map() and filter(), reduce() is in the functools module, so you must import it.

Syntax:

    from functools import reduce
    reduce(function, iterable[, initializer])

Example:
    
    from functools import reduce

    numbers = [1, 2, 3, 4]

    product = reduce(lambda x, y: x*y, numbers)
    
    print(product)   # 24  (i.e., (((1*2)*3)*4))

Use Case: Aggregating values (sum, product, max, etc.).

Example Using All Three Together
from functools import reduce

    numbers = [1, 2, 3, 4, 5, 6]

# Step 1: Square each number
squares = map(lambda x: x*x, numbers)         # [1, 4, 9, 16, 25, 36]

# Step 2: Keep only even squares
even_squares = filter(lambda x: x % 2 == 0, squares)  # [4, 16, 36]

product = reduce(lambda x, y: x*y, even_squares)       # 2304

```python
'''3.Write a Python function that checks if a given number is prime or not from
1 to 200.'''

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):  # check divisors up to sqrt(n)
        if n % i == 0:
            return False
    return True

# Check prime numbers from 1 to 200
for num in range(1, 201):
    if is_prime(num):
        print(num, "is Prime")
    else:
        print(num, "is Not Prime")
```

```python
'''5.  Create an iterator class in Python that generates the Fibonacci
sequence up to a specified number of terms.'''

class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms   # number of terms to generate
        self.count = 0               # track terms generated
        self.a, self.b = 0, 1        # starting values of Fibonacci

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration   # stop when max terms reached
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b


# Example usage:
fib = FibonacciIterator(10)  # generate 10 terms
for num in fib:
    print(num, end=" ")
```

```python
'''6. Write a generator function in Python that yields the powers of
2 up to a given exponent.'''

def powers_of_two(n):
    for i in range(n + 1):
        yield 2 ** i

# Example usage:
for value in powers_of_two(5):
    print(value, end=" ")

```

```python
'''10. Create a Python program that uses filter() to remove all the vowels
from a given string.'''

def remove_vowels(s):
    vowels = "aeiouAEIOU"
    return "".join(filter(lambda ch: ch not in vowels, s))

# Example usage:
text = "Hello World"
print(remove_vowels(text))
```
